```python
# function for Q3

def div3(a, b):
    # for range between a & b
    l = []
    i = a
    while i < b:
        if i%3 == 0:
            l.append(i)
        i += 1
    return l   
```

```python
div3(40, 80)
```

```python
# function vowel list:
def vowList(p):
    v= ["a","e","i","o","u"]
    l = []
    i = 0
    while i < len(p):
        j = 0
        while j < len(v):
            if p[i] == v[j]:
                l.append(p[i])
            j = j+1
        i +=1
    return l
```

```python
a = "This  is a test case for vowel list in this paragraph."

vowList(a)
```
