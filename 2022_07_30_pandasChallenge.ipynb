```python
'''We will have to change the datatypes for the analysis.'''
login['Login Time'] = pd.to_datetime(login['Login Time'], format='%H:%M:%S')
login['Logout Time'] = pd.to_datetime(login['Logout Time'], format='%H:%M:%S')
# Iam unable to directly change the DURATION TO timedelta type, so we created a new columns for analysis
login['Duration New'] = login['Logout Time'] - login['Login Time']

perform['Average Response Time'] = pd.to_datetime(perform['Average Response Time'], format='%H:%M:%S')
perform['Average Resolution Time'] = pd.to_datetime(perform['Average Resolution Time'], format='%H:%M:%S')
```

```python
#12. percentage of chat on which they have received a feedback

# making new columns of cummulative chat values

perform['CumChats'] = perform[['Agent Name', 'Total Chats']].groupby('Agent Name').cumsum()
```

```python
# column for cummulative Feedback value 
perform['CumFeeds'] = perform[['Agent Name', 'Total Feedback']].groupby('Agent Name').cumsum()
```

```python
perform.tail()
```

```python
# new column for the % value

for i in perform['SL No']:
    try:
        perform['pccf'] = (perform['CumFeeds']/perform['CumChats'])*100
    except:
        perform['pccf'] = 0
```

```python
perform[['Agent Name','pccf']].tail()
```

```python
perform.groupby('Agent Name')['pccf'].max()
```
